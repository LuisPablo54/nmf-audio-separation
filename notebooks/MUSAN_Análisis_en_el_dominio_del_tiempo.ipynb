```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("nhattruongdev/musan-noise")

print("Path to dataset files:", path)
```

```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import glob
import random
```

```python
categorias = ['speech', 'music', 'noise']
plt.figure(figsize=(15, 10))

for i, cat in enumerate(categorias):

    search_path = os.path.join(path, "musan", cat, "**", "*.wav")
    archivos = glob.glob(search_path, recursive=True)

    if archivos:
        archivo_random = random.choice(archivos)

        #5 segundos
        y, sr = librosa.load(archivo_random, duration=5)

        plt.subplot(3, 1, i+1)
        colores = {'speech': '#DE8C73', 'music': '#639ACA', 'noise': '#76B788'}

        librosa.display.waveshow(y, sr=sr, color=colores[cat])
        plt.title(f"Waveform de: {cat.upper()})", fontsize=14)
        plt.ylabel("Amplitud")
        if i < 2: plt.xlabel("")
    else:
        print(f"No se encontraron archivos en la ruta: {search_path}")

plt.tight_layout()
plt.show()
```
